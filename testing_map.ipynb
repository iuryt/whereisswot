```python
import os
import tempfile
from markupsafe import Markup  # Correct import for Markup
import geopandas as gpd
import pandas as pd
from datetime import timedelta, datetime, timezone
from io import BytesIO
import uuid
from tools import process_orbit_data, create_map  # Import create_map
import numpy as np
import os
# import hvplot.pandas
import folium
from folium.plugins import MousePosition
```

```python
orbit_data = process_orbit_data()
```

```python
# select orbit_data for TIME between Apr 25 and June 1st
start_date = np.datetime64('2025-04-25T00:00:00Z')
end_date = np.datetime64('2025-06-01T00:00:00Z')
selected_orbit_data = orbit_data[(orbit_data['TIME'] >= start_date) & (orbit_data['TIME'] <= end_date)]
selected_orbit_data.reset_index(drop=True, inplace=True)
```

```python
grouped = selected_orbit_data.groupby(selected_orbit_data['TIME'].dt.round("1D"))

for date, group in grouped:
    group.to_file(f"data/processed/selected_orbit_data/{date.date()}.kml", driver='KML')
```
